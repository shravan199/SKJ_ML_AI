```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#### 1. Age- Age of patient at time of operation (numerical)
#### 2. Year - Patient's year of operation (year - 1900, numerical)
#### 3. Num_Aux_Nodes- Number of positive axillary nodes detected (numerical)
#### 4. Survival status (class attribute)
    ##### 1 = the patient survived 5 years or longer
    ##### 2 = the patient died within 5 year
```

```python
haberman = pd.read_csv('./data/haberman.csv', names= ['age', 'year_operation', 'num_aux_nodes',  'survival_status'])
haberman
```

```python
# PDF (Prob. Density funciton) and CDF (Cumulative Distribution function)
```

```python
# plot cdf and pdf
counts, bin_edges = np.histogram(haberman['age'], bins= 10, density=True)


# compute pdf
pdf = counts/sum(counts)
print(pdf)
print(bin_edges)

# compute cdf
cdf = np.cumsum(pdf)

plt.plot(bin_edges[1: ], cdf)
plt.plot(bin_edges[1: ], pdf)
```
