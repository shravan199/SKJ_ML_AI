```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv('./data/iris.csv')
```

```python
iris.shape
```

```python
iris.head()
```

```python
iris['variety'].value_counts()
```

### 2-D scatter plot

```python
iris.plot(kind='scatter', x='sepal.length', y='sepal.width')
plt.show()
```

```python
plt.figure(figsize=(15, 6))

sns.set_style(style='whitegrid')
sns.FacetGrid(data=iris, hue='variety', height=4) \
   .map(plt.scatter, 'sepal.length', 'sepal.width') \
   .add_legend()

plt.show()
```

```python
iris_setosa = iris[(iris['sepal.width'] > 2.8)  & (iris['sepal.width'] < 4.5) &
                   (iris['sepal.length'] < 5.5)]
iris_setosa
```

```python
iris_setosa.info()
```

### Pair-plot

```python
plt.close()
sns.set_style(style='whitegrid')
sns.pairplot(data=iris, hue='variety', height=3)
plt.show()
```
