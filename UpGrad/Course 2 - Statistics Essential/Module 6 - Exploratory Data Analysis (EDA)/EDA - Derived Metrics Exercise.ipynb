```python
odi_centuries = odi_batting[odi_batting['Runs'] >= 100]
odi_centuries['Player'].value_counts()
```

```python
odi_centuries.groupby('Player')[['Player', 'Runs']].count()
```

```python
odi_centuries.pivot_table(index='Player', values='Runs').sort_values(by='Runs', ascending=False)
```

```python
odi_batting['Is_Century'] = odi_batting['Runs'].apply(lambda run: 1 if run >= 100 else 0)
```

```python
odi_batting.pivot_table(index='Player', values='Is_Century', aggfunc=np.sum).sort_values(by='Is_Century', ascending=False)
```

```python
odi_batting.groupby('Player')[['Player', 'Is_Century']].sum().sort_values(by='Is_Century', ascending=False)
```
