```python
kohli_df = pd.read_csv("../Cricket/Virat_Kohli_ODI.csv")
kohli_df
```

```python
greater_than_100_runs= [str(score) for score in np.arange(100, 201)]
#greater_than_100_runs[:5]
greater_than_100_runs_with_astrik= [ str(score) + '*' for score in np.arange(100, 201)]
series_1 = greater_than_100_runs_with_astrik.append(greater_than_100_runs)
print(series_1)
```

```python
filter_1 = kohli_df['Runs'].isin(greater_than_100_runs)
print(filter_1)
filter_2 = kohli_df['Runs'].isin(greater_than_100_runs_with_astrik)
print()
print(filter_2)
```

```python
kohli_df[filter_1 | filter_2].sort_values(by='Runs', ascending=True)
```
