```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tsd_345micensus = pd.read_csv("./Data/KPIsfrom3rd5thMICensus.csv")
tsd_345micensus
```

```python
plt.figure(figsize=(14,6))
plt.subplot(1, 3, 1)
sns.boxplot(y='3rd MI Census', data=tsd_345micensus)

plt.subplot(1, 3, 2)
sns.boxplot(y='4th MI Census', data=tsd_345micensus)

plt.subplot(1, 3, 3)
sns.boxplot(y='5th MI Census', data=tsd_345micensus)
plt.show()
```

```python
plt.figure(figsize=(16,6))
plt.subplot(1, 3, 1)
sns.distplot(tsd_345micensus['3rd MI Census'], rug=True)

plt.subplot(1, 3, 2)
sns.distplot(tsd_345micensus['4th MI Census'], rug=True)

plt.subplot(1, 3, 3)
sns.distplot(tsd_345micensus['5th MI Census'], rug=True)
plt.show()
```

```python
plt.figure(figsize=(14,6))
plt.subplot(1, 3, 1)
sns.scatterplot(x='3rd MI Census', y='Parameter', data=tsd_345micensus)

plt.subplot(1, 3, 2)
sns.scatterplot(x='4th MI Census', y='Parameter', data=tsd_345micensus)

plt.subplot(1, 3, 3)
sns.scatterplot(x='5th MI Census',y='Parameter',  data=tsd_345micensus)
plt.show()
```

```python
plt.figure(figsize=(20, 12))
plt.subplot(1, 3, 1)
sns.barplot(x='3rd MI Census', y='Parameter', data=tsd_345micensus)

plt.subplot(1, 3, 2)
sns.barplot(x='4th MI Census', y='Parameter', data=tsd_345micensus)


plt.show()
```
