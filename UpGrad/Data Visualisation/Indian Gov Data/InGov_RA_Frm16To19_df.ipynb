```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
RA_Frm16To19_df = pd.read_csv('./Data/RA2019_A2.csv')
RA_Frm16To19_df
```

```python
filter_criteria = RA_Frm16To19_df['States/UTs'].isin(['Total '])
RA_Frm16To19_df2 = RA_Frm16To19_df[~filter_criteria]
RA_Frm16To19_df2
```

```python
RA_Frm16To19_df2.info()
```

```python
RA_Frm16To19_df2.describe()
```

```python
plt.figure('1')
plt.figure(figsize=(14, 7))

plt.subplot(1, 4, 1)
sns.boxplot(y='State/UT-Wise Total Number of Road Accidents during 2016', data=RA_Frm16To19_df2)

plt.subplot(1, 4, 2)
sns.boxplot(y='State/UT-Wise Total Number of Road Accidents during 2017', data=RA_Frm16To19_df2)

plt.subplot(1, 4, 3)
sns.boxplot(y='State/UT-Wise Total Number of Road Accidents during 2018', data=RA_Frm16To19_df2)

plt.subplot(1, 4, 4)
sns.boxplot(y='State/UT-Wise Total Number of Road Accidents during 2019 - Numbers', data=RA_Frm16To19_df2)

plt.show()
```

```python
plt.figure('1')
plt.figure(figsize=(17, 7))

plt.subplot(1, 4, 1)
sns.histplot(x='State/UT-Wise Total Number of Road Accidents during 2016', data=RA_Frm16To19_df2)

plt.subplot(1, 4, 2)
sns.histplot(x='State/UT-Wise Total Number of Road Accidents during 2017', data=RA_Frm16To19_df2)

plt.subplot(1, 4, 3)
sns.histplot(x='State/UT-Wise Total Number of Road Accidents during 2018', data=RA_Frm16To19_df2)

plt.subplot(1, 4, 4)
sns.histplot(x='State/UT-Wise Total Number of Road Accidents during 2019 - Numbers', data=RA_Frm16To19_df2)

plt.show()
```
